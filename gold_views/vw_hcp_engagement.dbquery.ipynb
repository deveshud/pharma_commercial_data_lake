```sql
CREATE OR REPLACE VIEW cdl.gold.vw_hcp_engagement AS
SELECT
    hcp_id,
    hcp_name,
    specialty,
    city,
    state,
    territory_id,
    territory_name,
    region,
    zone,
    sales_rep_id,
    sales_rep_name,

    total_calls,
    completed_calls,
    planned_calls,
    cancelled_calls,
    avg_call_duration,
    last_call_date,

    total_prescriptions,
    last_rx_date,

    ROUND(
        completed_calls / NULLIF(total_calls, 0),
        2
    ) AS call_completion_rate,

    CASE
        WHEN total_prescriptions >= 50 AND total_calls < 3 THEN 'High RX / Low Call'
        WHEN total_prescriptions >= 50 AND total_calls >= 3 THEN 'High Value Engaged'
        WHEN total_prescriptions < 50 AND total_calls >= 3 THEN 'Engaged / Low RX'
        ELSE 'Low Engagement'
    END AS hcp_segment,

    gold_load_ts
FROM cdl.gold.hcp_engagement;
```
