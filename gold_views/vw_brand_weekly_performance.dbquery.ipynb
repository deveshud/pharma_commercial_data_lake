```sql
CREATE OR REPLACE VIEW cdl.gold.vw_brand_weekly_performance AS
SELECT
    brand,
    therapy_area,
    week_start_date,

    total_prescriptions,
    total_hcps_prescribing,
    total_calls,
    completed_calls,

    ROUND(
        completed_calls / NULLIF(total_calls, 0),
        2
    ) AS call_completion_rate,

    ROUND(
        total_prescriptions / NULLIF(total_hcps_prescribing, 0),
        2
    ) AS rx_per_prescribing_hcp,

    gold_load_ts
FROM cdl.gold.brand_weekly_performance;

```
