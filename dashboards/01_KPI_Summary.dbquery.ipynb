```sql
SELECT
    SUM(total_prescriptions) AS total_prescriptions,
    SUM(total_calls) AS total_calls,
    SUM(completed_calls) AS completed_calls,

    ROUND(
        SUM(completed_calls) / NULLIF(SUM(total_calls), 0),
        2
    ) AS overall_call_completion_rate,

    COUNT(DISTINCT brand) AS total_brands,
    COUNT(DISTINCT therapy_area) AS total_therapy_areas,

    MAX(week_start_date) AS latest_week_start_date
FROM cdl.gold.vw_brand_weekly_performance;
```
