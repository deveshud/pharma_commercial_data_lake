```python
from pyspark.sql import Row

employees_data = [
    Row(emp_id=1, emp_name="Amit", dept_id=10, salary=70000),
    Row(emp_id=2, emp_name="Neha", dept_id=20, salary=85000),
    Row(emp_id=3, emp_name="Ravi", dept_id=10, salary=60000),
    Row(emp_id=4, emp_name="Priya", dept_id=30, salary=90000),
    Row(emp_id=5, emp_name="Karan", dept_id=20, salary=75000)
]

departments_data = [
    Row(dept_id=10, dept_name="Engineering"),
    Row(dept_id=20, dept_name="Analytics"),
    Row(dept_id=30, dept_name="Finance")
]

employees_df = spark.createDataFrame(employees_data)
departments_df = spark.createDataFrame(departments_data)

display(employees_df)
display(departments_df)
```

```python
from pyspark.sql.functions import sum

df_dept_total_sal = departments_df.alias("dept").join(employees_df,on = "dept_id",how = "left").groupBy("dept_name").agg(sum("salary").alias("total_salary"))

display(df_dept_total_sal)
```
